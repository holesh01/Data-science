```python
sns.boxplot(x='RL',y='LotFrontage',data=house)
```

```python
def add_LotFrontage(x):
    if(pd.isnull(x[0])):
        if(x[1]==0):
            return 57

        else :
            return 75
    else:
        return x[0]
        
```

```python
house['LotFrontage']=house[['LotFrontage','RL']].apply(add_LotFrontage,axis=1)
```

```python
w4=pd.get_dummies(house['LandContour'],drop_first=True)
s1=pd.get_dummies(house['LotShape'],drop_first=True)
s2=pd.get_dummies(house['Utilities'],drop_first=True)
s3=pd.get_dummies(house['LotConfig'],drop_first=True)
s4=pd.get_dummies(house['LandSlope'],drop_first=True)
s6=pd.get_dummies(house['Neighborhood'],drop_first=True)
a1=pd.get_dummies(house['Condition1'],drop_first=True)
a2=pd.get_dummies(house['Condition2'],drop_first=True)
a3=pd.get_dummies(house['BldgType'],drop_first=True)
a4=pd.get_dummies(house['HouseStyle'],drop_first=True)
a5=pd.get_dummies(house['RoofStyle'],drop_first=True)
q1=pd.get_dummies(house['Exterior1st'],drop_first=True)
q2=pd.get_dummies(house['ExterCond'],drop_first=True)
q3=pd.get_dummies(house['Foundation'],drop_first=True)
q4=pd.get_dummies(house['Heating'],drop_first=True)
w1=pd.get_dummies(house['GarageCond'],drop_first=True)
w2=pd.get_dummies(house['SaleCondition'],drop_first=True)
w3=pd.get_dummies(house['SaleType'],drop_first=True)
```

```python
house=pd.concat([house,s1,s2,s3,s4,s6,a1,a2,a3,a4,a5,q1,q2,q3,w1,w2,w3,w4],axis=1)
```
